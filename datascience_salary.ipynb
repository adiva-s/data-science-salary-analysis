```python
import pandas as pd
```

```python
salary = pd.read_csv('Data Science Salary 2021 to 2023.csv')
salary.head()
```

```python
salary.info()
```

```python
salary['salary'].describe()
```

```python
salary['job_title'].value_counts()
```

```python
drop_columns = ['salary_currency', 'salary']
salary = salary.drop(
    labels=drop_columns,
    axis=1
)
```

```python
salary.head()
```

```python
column_mapper = {'salary_in_usd':'converted_salary_usd'}
salary = salary.rename (mapper=column_mapper, axis=1)
salary.head()
```

```python
# Define the list of desired job title - first case - Data Scientist
desired_job_title = ["Data Scientist"]

# Filter the DataFrame to include only the desired job title
filtered = salary[salary['job_title'].isin(desired_job_title)]

# Sort the filtered DataFrame by the "work_year" column with ascending = True
sorted_df = filtered.sort_values(
    by='work_year',
    ascending = True
)

print(sorted_df)
```

```python
# Define the list of desired job title - second case - Data Engineer
desired_job_title = ["Data Engineer"]

# Filter the DataFrame to include only the desired job title
filtered = salary[salary['job_title'].isin(desired_job_title)]

# Sort the filtered DataFrame by the "work_year" column with ascending = True
sorted_df = filtered.sort_values(
    by='work_year',
    ascending = True
)

print(sorted_df)
```
